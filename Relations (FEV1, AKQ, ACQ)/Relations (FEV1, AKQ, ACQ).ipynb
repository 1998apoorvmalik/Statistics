```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './data.xlsx'
df = pd.read_excel(path)
df.head()
```

```python
group_a = np.array([np.array(df.iloc[:, 0]), np.array(df.iloc[:, 1]), np.array(df.iloc[:, 2])])
group_b = np.array([np.array(df.iloc[:, 4]), np.array(df.iloc[:, 5]), np.array(df.iloc[:, 6])])
group_c = np.array([np.array(df.iloc[:, 8]), np.array(df.iloc[:, 9]), np.array(df.iloc[:, 10])])
```

```python
#Plotting FEV1 vs ACQ:
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10, 25))
min_xlim = min(min(group_a[0]), min(group_b[0]), min(group_c[0]))-0.1
max_xlim = max(max(group_a[0]), max(group_b[0]), max(group_c[0]))+0.1

ax1.scatter(group_a[0], group_a[2], color='green')
ax1.set_xlim(min_xlim, max_xlim)
ax1.set_title("Group A")
ax1.set_xlabel("FEV1 --->")
ax1.set_ylabel("ACQ --->")
a_tx = max([group_a[0][i] for i in range(len(group_a[0])) if group_a[2][i]==0])
ax1.axvline(x=a_tx, color='red', linestyle='dashed', label="x = "+str(a_tx))
ax1.annotate('Threshold', xy=(a_tx, -0.05), xytext=(a_tx, -0.05), arrowprops=dict(facecolor='pink'))
ax1.legend(loc=2)

ax2.scatter(group_b[0], group_b[2], color='green')
ax2.set_xlim(min_xlim, max_xlim)
ax2.set_title("Group B")
ax2.set_xlabel("FEV1 --->")
ax2.set_ylabel("ACQ --->")
b_tx = max([group_b[0][i] for i in range(len(group_b[0])) if group_b[2][i]==0])
ax2.axvline(x=b_tx, color='red', linestyle='dashed', label="x = "+str(b_tx))
ax2.annotate('Threshold', xy=(b_tx, -0.05), xytext=(b_tx, -0.05), arrowprops = dict(facecolor='pink'))
ax2.legend(loc=2)

ax3.scatter(group_c[0], group_c[2], color='green')
ax3.set_xlim(min_xlim, max_xlim)
ax3.set_title("Group C")
ax3.set_xlabel("FEV1 --->")
ax3.set_ylabel("ACQ --->")
c_tx = max([group_c[0][i] for i in range(len(group_c[0])) if group_c[2][i]==0])
ax3.axvline(x=c_tx, color='red', linestyle='dashed', label="x = "+str(c_tx))
ax3.annotate('Threshold', xy=(c_tx, -0.05), xytext=(c_tx, -0.05), arrowprops = dict(facecolor='pink'))
ax3.legend(loc=2)
```
